```python
import numpy as np
import pandas as pd

data = pd.read_csv(r"D:\Oladimeji\Data Science\DataSet\Kaggle\Healthcare Providers.csv")
```

```python
# Checking the number of rows and columns to clean
data.shape
```

```python
# To get the overview of the columns in the data
data.info()
```

```python
# To get the sum of all the missing values in the columns
data.isna().sum()
```

```python
# Dropping the column because Street Address 1 contains more informations needed
data.drop(columns = ["Street Address 2 of the Provider"], inplace = True)
```

```python
data.isna().sum()
```

```python
# Getting the firt 5 rows from the data table
data.head()
```

```python
# Getting the number of the missing value in the provider gender column
data["Gender of the Provider"].value_counts()
```

```python
# To replace the null values with unknown to clean the gender column
data.fillna({"Gender of the Provider":"UNKNOWN"}, inplace = True)
```

```python
# Doing the value count to confirm the changes
data["Gender of the Provider"].value_counts()
```

```python
# Cofiming the changes on the original data
data.isna().sum()
```

```python
# Filling all missing values in the initial columns with 'No Initials'. It can not be guessed
data.fillna({"Middle Initial of the Provider": "No Initial"}, inplace = True)
```

```python
data.isna().sum()
```

```python
# Replacing the missing name of the provider with the national provider identification because it is a unique number
data["First Name of the Provider"] = data["First Name of the Provider"].fillna(data["National Provider Identifier"].astype(str))
```

```python
# Looking at the remaining column with missing values
data.isna().sum()
```

```python
# The first ten information in the variable "data"
data.head(10)
```

```python
# Replacing the missing Credentials of the Provider with Not Provided because it cannot be guessed
data["Credentials of the Provider"] = data["Credentials of the Provider"].fillna("Not Provided")
```

```python
# Doing the final checks if there is any missing values
data.isna().sum()
```

```python
data.info()
```

```python
# Saving columns index 23, 24 and 25 into a variable name avg_amt
avg_amt = ["Average Medicare Allowed Amount", "Average Submitted Charge Amount", "Average Medicare Payment Amount"]
```

```python
# Converting the data type of columns 23, 24, and 25 from object to float
data[avg_amt] = data[avg_amt].replace(',', '', regex=True).astype(float)
```

```python
data.info()
```

```python
# Converting the data type of column 'Average Medicare Standardized Amount' from object to float
data["Average Medicare Standardized Amount"] = data["Average Medicare Standardized Amount"].replace(',', '', regex=True).astype(float)
```

```python
# Confirming the conversions

data.info()
```

```python
# Saving columns 19, 20, and 21 into the variable name num_of_serv so as to convert them at once
num_of_serv = ["Number of Services", "Number of Medicare Beneficiaries", "Number of Distinct Medicare Beneficiary/Per Day Services"]
```

```python
# Converting the variable from object data type to float data type
data[num_of_serv] = data[num_of_serv].replace(',', '', regex=True).astype(float)
```

```python
# Converting the variable from float data type to integer data type
data[num_of_serv] = data[num_of_serv].astype(int)
```

```python
# Doing the final confirmation
data.info()
```
